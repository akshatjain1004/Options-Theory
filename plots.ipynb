Models

```python
#binomial model
def binomial_model(S, K, T, r, sigma, N, option_type):
    """
    This function calculates the price of an option using the binomial model.

    Parameters:
        S (float): current stock price
        K (float): strike price
        T (float): time to maturity (in years)
        r (float): risk-free rate
        sigma (float): volatility
        N (int): number of time steps
        option_type (str): "European" or "American"

    Returns:
        option_price (float): price of the option
    """

    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    stock_price = np.zeros((N+1, N+1))
    option_value = np.zeros((N+1, N+1))

    for i in range(N+1):
        for j in range(i+1):
            stock_price[j, i] = S * (u ** (i-j)) * (d ** j)

    if option_type == "European":
        option_value[:, N] = np.maximum(0, K - stock_price[:, N])
    elif option_type == "American":
        option_value[:, N] = np.maximum(0, K - stock_price[:, N])

    for i in range(N-1, -1, -1):
        for j in range(i+1):
            if option_type == "European":
                option_value[j, i] = np.exp(-r * dt) * (p * option_value[j, i+1] + (1-p) * option_value[j+1, i+1])
            elif option_type == "American":
                exercise_value = K - stock_price[j, i]
                option_value[j, i] = max(exercise_value, np.exp(-r * dt) * (p * option_value[j, i+1] + (1-p) * option_value[j+1, i+1]))

    stock_price = stock_price[0, 0]
    option_price = option_value[0, 0]

    return option_price, stock_price


#black scholes model
def black_scholes(S, K, T, r, sigma, option_type):
    """
    This function calculates the price of an option using the Black-Scholes formula.

    Parameters:
        S (float): current stock price
        K (float): strike price
        T (float): time to maturity (in years)
        r (float): risk-free rate
        sigma (float): volatility
        option_type (str): "call" or "put"

    Returns:
        option_price (float): price of the option
    """

    d1 = (np.log(S/K) + (r + sigma**2/2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == "call":
        option_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == "put":
        # option_price = K * np.exp(-r * T) * norm.cdf(-d2)
        option_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

    return option_price

#delta strategy
def delta_strategy(S, K, T, r, sigma, N, option_type="European"):
    """
    This function calculates the price of an option using the binomial model.

    Parameters:
        S (float): current stock price
        K (float): strike price
        T (float): time to maturity (in years)
        r (float): risk-free rate
        sigma (float): volatility
        N (int): number of time steps
        option_type (str): "European" (self-financing strategy is only available for European options)

    Returns:
        weekly_deltas (list): weekly deltas for the self-financing strategy
    """

    dt = T / N
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    stock_price = np.zeros((N+1, N+1))
    option_value = np.zeros((N+1, N+1))
    deltas = np.zeros((N, N))

    for i in range(N+1):
        for j in range(i+1):
            stock_price[j, i] = S * (u ** (i-j)) * (d ** j)

    if option_type == "European":
        option_value[:, N] = np.maximum(0, K - stock_price[:, N])

    for i in range(N-1, -1, -1):
        for j in range(i+1):
            if option_type == "European":
                option_value[j, i] = np.exp(-r * dt) * (p * option_value[j, i+1] + (1-p) * option_value[j+1, i+1])
                numerator   = option_value[j, i+1] - option_value[j+1, i+1]
                denominator = stock_price[j, i+1] - option_value[j+1, i+1]
                deltas[j, i] = numerator / denominator

    stock_price = option_value[0, 0]
    option_price = option_value[0, 0]
    delta_initial = deltas[0, 0]

    return option_price, stock_price, delta_initial
```

```python
# Calculate the self-financing strategy for each stock
strategy = pd.DataFrame(columns=["Week", "Stock", "Delta", "Stock Price", "Option Price", "Holding", "Cash", "Total Value"])

# Loop through each stock and strike price
for stock in stocks.keys():
    for K in K_list:

        # Calculate the delta value using the Black-Scholes formula
        delta = black_scholes_delta(
            S=stocks[stock]["S"],
            K=K,
            T=stocks[stock]["T"],
            r=stocks[stock]["r"],
            sigma=implied_volatility,
            option_type="put"
        )

        # Initialize the variables for the self-financing strategy
        stock_price = stocks[stock]["S"]
        option_price = black_scholes_price(
            S=stocks[stock]["S"],
            K=K,
            T=stocks[stock]["T"],
            r=stocks[stock]["r"],
            sigma=implied_volatility,
            option_type="put"
        )
        holding = -delta
        cash = delta * stock_price + option_price
        total_value = holding * stock_price + cash

        # Add the results to the DataFrame for the first week
        strategy = strategy.append({
            "Week": 0,
            "Stock": stock,
            "Delta": delta,
            "Stock Price": stock_price,
            "Option Price": option_price,
            "Holding": holding,
            "Cash": cash,
            "Total Value": total_value
        }, ignore_index=True)

        # Loop through each week
        for week in range(1, 53):

            # Simulate the change in the stock price
            stock_return = np.random.normal(
                loc=stocks[stock]["r"],
                scale=stocks[stock]["sigma"]
            )
            stock_price = stock_price * np.exp(stock_return)

            # Calculate the new option price using the Black-Scholes formula
            option_price = black_scholes_price(
                S=stock_price,
                K=K,
                T=stocks[stock]["T"] - week / 52,
                r=stocks[stock]["r"],
                sigma=implied_volatility,
                option_type="put"
            )

            # Calculate the new delta value using the Black-Scholes formula
            delta = black_scholes_delta(
                S=stock_price,
                K=K,
                T=stocks[stock]["T"] - week / 52,
                r=stocks[stock]["r"],
                sigma=implied_volatility,
                option_type="put"
            )

            # Calculate the new holding and cash values for the self-financing strategy
            new_holding = -delta
            new_cash = cash * np.exp(stocks[stock]["r"] / 52) + (holding - new_holding) * stock_price
            new_total_value = new_holding * stock_price + new_cash

            # Add the results to the DataFrame for the current week
            strategy = strategy.append({
                "Week": week,
                "Stock": stock,
                "Delta": delta,
                "Stock Price": stock_price,
                "Option Price": option_price,
                "Holding": new_holding,
                "Cash
                "Cash": new_cash,
                "Total Value": new_total_value
            }, ignore_index=True)

            # Update the holding and cash values for the next week
            holding = new_holding
            cash = new_cash

# Plot the results for the self-financing strategy
import matplotlib.pyplot as plt

fig, axs = plt.subplots(len(stocks) * len(K_list), figsize=(10, 15), sharex=True)

for i, stock in enumerate(stocks.keys()):
    for j, K in enumerate(K_list):
        df = strategy[(strategy["Stock"] == stock) & (strategy["Option Price"] == K)]
        axs[i*len(K_list) + j].plot(df["Week"], df["Total Value"])
        axs[i*len(K_list) + j].set_title(f"{stock} - Strike Price {K}")
        axs[i*len(K_list) + j].set_xlabel("Week")
        axs[i*len(K_list) + j].set_ylabel("Total Value")

plt.tight_layout()
plt.show()
```

```python
def delta_strategy(S, K, T, r, option_type):
    """
    This function calculates the delta and constructs a self-financing strategy using the Black-Scholes model
    with implied volatility.

    Parameters:
        S (float): current stock price
        K (float): strike price
        T (float): time to maturity (in years)
        r (float): risk-free rate
        option_type (str): "call" or "put"

    Returns:
        delta_values (list): list of delta values
        strategy (list): list of self-financing strategy values
    """

    sigma_imp = black_scholes_implied_volatility(S, K, T, r, option_type)
    delta_values = []
    strategy = []

    # Initial values
    delta_t0 = norm.cdf(d1(S, K, T, r, sigma_imp))
    strategy_t0 = delta_t0 * S - option_price(S, K, T, r, sigma_imp, option_type)
    delta_values.append(delta_t0)
    strategy.append(strategy_t0)

    # Calculate delta and self-financing strategy for each week
    for i in range(1, int(T*52)+1):
        t = i / 52
        S_i = S * np.exp(r * t)
        sigma_imp_i = black_scholes_implied_volatility(S_i, K, T-t, r, option_type)
        delta_i = norm.cdf(d1(S_i, K, T-t, r, sigma_imp_i))
        delta_values.append(delta_i)
        strategy_i = strategy[i-1] * np.exp(r/52) + (delta_i - delta_values[i-1]) * S_i
        strategy.append(strategy_i)

    return delta_values, strategy

def black_scholes_implied_volatility(S, K, T, r, option_type, option_price):
    """
    This function calculates the implied volatility using the Black-Scholes formula.

    Parameters:
        S (float): current stock price
        K (float): strike price
        T (float): time to maturity (in years)
        r (float): risk-free rate
        option_type (str): "call" or "put"
        option_price (float): price of the option

    Returns:
        sigma_imp (float): implied volatility
    """

    tol = 1e-6
    max_iter = 1000
    sigma = 0.5

    for i in range(max_iter):
        option_price_calculated = black_scholes(S, K, T, r, sigma, option_type)
        vega = S * norm.pdf(d1(S, K, T, r, sigma)) * np.sqrt(T)
        diff = option_price_calculated - option_price

        if abs(diff) < tol:
            return sigma

        sigma = sigma - diff / vega

    return None
```
